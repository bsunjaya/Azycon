```python
station_to_resort = {
    71032: "Thredbo AWS",
    71075: "Perisher AWS",
    72161: "Cabramurra SMHEA AWS",
    83024: "Mount Buller",
    83084: "Falls Creek",
    83085: "Mount Hotham",
    85291: "Mount Baw Baw"
}
# Add new column
df_2014["Resort"] = df_2014["Bureau of Meteorology station number"].map(station_to_resort)
df_2014.head()
```

# Feature Engineering

```python
df_2014["Date"] = pd.to_datetime(df_2014[["Year", "Month", "Day"]])
df_2014.head()
```

```python
df_2014["Mean_Temp"]=  (df_2014["Maximum temperature (Degree C)"] + df_2014["Minimum temperature (Degree C)"]) / 2
```

```python
# Function to convert rainfall (mm) into snowfall (cm)
def rainfall_to_snow(row):
    t = row["Mean_Temp"]
    rain_mm = row["Rainfall amount (millimetres)"]
    
    if pd.isna(rain_mm):  # handle missing values
        return 0
    
    if t <= 0:
        slr = 10   # normal snow (10:1)
    elif 0 < t < 2:
        slr = 5    # wet, dense snow (5:1)
    else:
        return 0   # too warm, all rain (just water) 
    
    # Convert mm water to cm snow using ratio
    return rain_mm * slr / 10

# Apply conversion to dataset
df_2014["Snowfall_cm"] = df_2014.apply(rainfall_to_snow, axis=1)
```

```python
# Temperature factor based on max temp (logistic decay)
df_2014["Snow_Retention"] = 1 / (1 + np.exp((df_2014["Maximum temperature (Degree C)"] - 0) / 2))
df_2014["Snow_Consistency"] = 1 / (1 + np.exp((df_2014["Minimum temperature (Degree C)"] - 0) / 2))
df_2014.tail()
```

```python
# Convert Date column to datetime if not already
df_2014["Date"] = pd.to_datetime(df_2014["Date"])

# Aggregate daily → weekly per resort
df_weekly = (
    df_2014.groupby(["Resort", pd.Grouper(key="Date", freq="W-SUN")])
    .agg({
        "Maximum temperature (Degree C)": "mean",   # weekly avg temp
        "Minimum temperature (Degree C)": "mean",   # weekly avg temp
        "Rainfall amount (millimetres)": "sum",     # weekly total rainfall
        "Snowfall_cm": "sum",                        # weekly snow total
        "Snow_Retention": "mean",
        "Snow_Consistency": "mean"
        
    })
    .reset_index()
    .rename(columns={
        "Maximum temperature (Degree C)": "TempMax",
        "Minimum temperature (Degree C)": "TempMin",
        "Rainfall amount (millimetres)": "Rainfall"
    })
)

# Inspect result
print(df_weekly.head())
```

```python
df_weekly.head()
```

```python
df_weekly["Snow_Availability"] = df_weekly["Snowfall_cm"] / df_weekly["Snowfall_cm"].max()
# Snow Reliability Index
w_availability = 0.5
w_retention = 0.3
w_consistency = 0.2

df_weekly["SRI"] = (
    w_availability * df_weekly["Snow_Availability"] +
    w_retention * df_weekly["Snow_Retention"] +
    w_consistency * df_weekly["Snow_Consistency"]
)
```

## Two Stage Hybrid RF Model

### SARIMA

```python
def forecast_weather(series, steps=75, var_name="Unknown"):
    """SARIMA forecast for a weather variable, fallback to mean if fails"""
    try:
        model = SARIMAX(series, order=(1,1,1), seasonal_order=(1,1,1,52),
                        enforce_stationarity=False, enforce_invertibility=False)
        result = model.fit(disp=False)
        forecast = result.get_forecast(steps=steps).predicted_mean
        print(f"SARIMA success: {var_name}")
        return forecast
    except Exception as e:
        print(f"SARIMA failed for {var_name}, using mean instead. Error: {e}")
        return pd.Series([series.mean()]*steps, 
                         index=pd.date_range(start=series.index[-1]+pd.Timedelta(weeks=1), 
                                             periods=steps, freq="W-SUN"))
```

```python
def hybrid_forecast_resort(df_resort, resort_name, steps=75):
    print(f"\n=== Processing {resort_name} ===")
    df_resort = df_resort.set_index("Date").asfreq("W-SUN").fillna(0)

    # ---- Stage 1: Forecast weather ----
    future_dates = pd.date_range(start=df_resort.index.max() + pd.Timedelta(weeks=1),
                                 periods=steps, freq="W-SUN")

    snow_forecast = forecast_weather(df_resort["Snowfall_cm"], steps, "Snowfall")
    tmax_forecast = forecast_weather(df_resort["TempMax"], steps, "TempMax")
    tmin_forecast = forecast_weather(df_resort["TempMin"], steps, "TempMin")
    rain_forecast = forecast_weather(df_resort["Rainfall"], steps, "Rainfall")

    future_weather = pd.DataFrame({
        "Date": future_dates,
        "Snowfall_cm": snow_forecast.values,
        "TempMax": tmax_forecast.values,
        "TempMin": tmin_forecast.values,
        "Rainfall": rain_forecast.values
    }).set_index("Date")

    # ---- Stage 2: Train RF ----
    df_resort["Week"] = df_resort.index.isocalendar().week.astype(int)
    df_resort["sin_week"] = np.sin(2*np.pi*df_resort["Week"]/52)
    df_resort["cos_week"] = np.cos(2*np.pi*df_resort["Week"]/52)
    df_resort["SRI_lag1"] = df_resort["SRI"].shift(1)
    df_resort["SRI_lag2"] = df_resort["SRI"].shift(2)
    df_resort = df_resort.dropna()

    features = ["SRI_lag1","SRI_lag2","sin_week","cos_week",
                "Snowfall_cm","TempMax","TempMin","Rainfall",
                "Snow_Retention","Snow_Consistency","Snow_Availability"]

    X_train, y_train = df_resort[features], df_resort["SRI"]

    rf = RandomForestRegressor(n_estimators=200, random_state=42, n_jobs=-1)
    rf.fit(X_train, y_train)
    print("RF training done")

    # ---- Stage 2b: Recursive forecast ----
    future_weather["Week"] = future_weather.index.isocalendar().week.astype(int)
    future_weather["sin_week"] = np.sin(2*np.pi*future_weather["Week"]/52)
    future_weather["cos_week"] = np.cos(2*np.pi*future_weather["Week"]/52)

    last_sri1 = df_resort.iloc[-1]["SRI"]
    last_sri2 = df_resort.iloc[-2]["SRI"]
    preds = []

    for d, row in future_weather.iterrows():
        feat = {
            "SRI_lag1": last_sri1,
            "SRI_lag2": last_sri2,
            "sin_week": row["sin_week"],
            "cos_week": row["cos_week"],
            "Snowfall_cm": row["Snowfall_cm"],
            "TempMax": row["TempMax"],
            "TempMin": row["TempMin"],
            "Rainfall": row["Rainfall"],
            "Snow_Retention": 0,   
            "Snow_Consistency": 0,
            "Snow_Availability": 0
        }
        feat = pd.DataFrame([feat])
        yhat = rf.predict(feat)[0]
        preds.append(yhat)

        # update lags
        last_sri2 = last_sri1
        last_sri1 = yhat

    fcast = pd.Series(preds, index=future_dates, name="Hybrid_SRI")
    print(f"Forecast complete for {resort_name}")
    return fcast
```

```python
predictions_2026_hybrid = {}
steps_ahead = 75

for resort in df_weekly["Resort"].unique():
    df_resort = df_weekly[df_weekly["Resort"] == resort].copy()
    fcast = hybrid_forecast_resort(df_resort, resort, steps=steps_ahead)
    predictions_2026_hybrid[resort] = fcast
```
